```python
from bs4 import BeautifulSoup
import selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import numpy as np
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import re
print(selenium.__version__)
```

```python
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
match_list = []
for page in range(3, 7):
    columns = ['contest_name', 'play_date', 'event', 'play_category', 'play_level', 'player_name', 'team_name', 'birth_date', 
               'weight/height', 'specialized_technique', 'first_learning_year', 'pro_debut_year', 'winner', 'match_result', 'text_relay']
    values = []
    url = 'https://kpso.co.kr/ksa/Schedule/GameScheduleAndResult?page_no={}&tab=all&season_id=2023&comp_id=25&section_cd=0&gender_cd=0&weight_cd=0&t_id=0'.format(page)
    driver.get(url)
    time.sleep(0.05)
    try:
        driver.maximize_window()
    except:
        pass
    
    print('{}페이지 크롤링'.format(page))
    soup = BeautifulSoup(driver.page_source, 'lxml')
    result_list = soup.find('tbody').find_all('tr')
    for i, result in enumerate(result_list):
        weight = result.find_all('td')
        if weight[3].get_text().strip() != '금강':
            continue
        print('{} 금강장사 경기 크롤링'.format(weight[6].get_text().strip()))
        row_list = []
        row_list.append(weight[0].get_text().strip())
        row_list.append(weight[1].get_text().strip())
        row_list.append(weight[2].get_text().strip())
        row_list.append(weight[3].get_text().strip())
        row_list.append(weight[6].get_text().strip())
        
        live_center_xpath = driver.find_element(By.XPATH, '//*[@id="{}"]/td[8]/a'.format(int(result['id'])))
        live_center_xpath.click()
        time.sleep(0.05)

        soup = BeautifulSoup(driver.page_source, 'lxml')
        blue_pic_box = soup.find_all('div', attrs={'class':'pic-box'})[0]
        red_pic_box = soup.find_all('div', attrs={'class':'pic-box'})[1]
        blue_name = blue_pic_box.find_all('p')
        red_name = red_pic_box.find_all('p')
        for blue, red in zip(blue_name, red_name):
            blue = blue.get_text().strip()
            red = red.get_text().strip()
            row_list.append((blue, red))


        profile_box = soup.find('div', attrs={'class':'profile-box'})
        blue_player = profile_box.find('div', attrs={'class':'blue'})
        blue_player_profile = blue_player.find_all('li')
        red_player = profile_box.find('div', attrs={'class':'red'})
        red_player_profile = red_player.find_all('li')
        for blue, red in zip(blue_player_profile, red_player_profile):
            blue = blue.get_text().strip()
            red = red.get_text().strip()
            row_list.append((blue, red))

        winner = soup.find('div', attrs={'class':'summary-box'})
        winner_text = winner.find('li').get_text().strip()
        final_text = winner_text[:3] + ' ' + winner_text[4] + ' ' + winner_text[5:]
        row_list.append(final_text)

        match_result = winner.find_all('li')[1].find_all('p')[1].get_text().strip()
        row_list.append(match_result)

        text_relay_xpath = driver.find_element(By.XPATH, '//*[@id="{}"]/div[2]/dl/dd[3]/a'.format(int(result['id'])))
        text_relay_xpath.click()
        time.sleep(0.05)
        
        soup = BeautifulSoup(driver.page_source, 'lxml')
        text_relay = soup.find('div', attrs={'class':'tab-box live'})
        text_list = text_relay.find_all('p')
        sentence = ''
        for text in text_list:
            sentence += (text.get_text().strip() + '\n')
        sentence = sentence.replace('영상보기', '')
        sentence = sentence.replace('* 문자중계 중계 선택 시 해당 영상이 표출됩니다.', '')
        sentence = sentence.strip()
        row_list.append(sentence)
        # print(row_list)
        values.append(row_list)
        
        driver.back()
        time.sleep(0.05)
        driver.back()
        time.sleep(0.05)

    df = pd.DataFrame(values, columns = columns)
    match_list.append(df)
    

    
```

```python
df_ = pd.concat(match_list).reset_index(drop=True) # index 초기화 해줘야함
df_.head()
```
